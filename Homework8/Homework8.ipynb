```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


##1
def make_sine_wave(x, A, w):
    return A * np.sin(w*x)

x = np.linspace(0, 2 * np.pi, 1000)
amp = [0.5,1,1.5,2,2.5]
freq = [1,2,3,4,5]
pars = zip(amp, freq)
y = []

listp = pars
listp = np.array(listp)
for i in range(0, len(list(pars))):
    y.append(make_sine_wave(x, amp[i], freq[i]))

y = np.array(y)
plt.figure(figsize = (15,4))

for i in range(0, len(y)):
    plt.plot(x, y[i], label = "A, w = ")

plt.grid()
plt.title("Sine Wave Abomination")
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.legend()

list(listp)
```
